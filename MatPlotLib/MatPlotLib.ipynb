### Basic Graph

```python
plt.plot([1, 2, 3], [2, 4, 6])
```

```python
# a good practice is to put the coordinates in a variable rather than directly typing it to the method itself as done above

x = [0, 1, 2, 3, 4]
y = [0, 2, 4, 6, 8]

#resize your graph (dpi specifies pixels per inch when saving probably should use 300 if possible)
plt.figure(figsize=(5, 3), dpi = 200) 

# Line 1
#keyword argument notation
#plt.plot(x, y, label='2x', color='red', linewidth = 3, marker='.', linestyle="--", markersize = 10, markeredgecolor='blue')

# short hand for the above line
plt.plot(x, y, 'b^--', label='2x') # format = [color][marker][line]

# Line 2

# select intervals we want to plot the points at
x2 = np.arange(0, 4.5, 0.5)

# to make the curve appear like its continuing in that trend.
plt.plot(x2[:8], x2[:8]**2, 'r', label='x^2')
plt.plot(x2[7:], x2[7:]**2, 'r--')

# specify font parameters with fontdict
plt.title("Our first Graph!", fontdict={'fontname': "Comic Sans MS", 'fontsize' : 20})

# X and Y labels
plt.xlabel('X axis')
plt.ylabel('Y axis')

# X and Y tickmarks(scale of the grpah)
plt.xticks([0, 1, 2, 3, 4, ])
plt.yticks([0, 2, 4, 6, 8, 10])

# Add a legend
plt.legend()

#Save figure(dpi 300 is good when saving so graph has high resolution)
plt.savefig('mygraph.png', dpi=300)

#show plot
plt.show() # to remove the line that came above the graph in the prev. cell
```

## Bar chart

```python
#all the attributes used for the line graph can also be used here
labels = ['A', 'B', 'C']
values = [1, 4, 2]
bars = plt.bar(labels, values)

patterns = ['/', 'O', '*']
for bar in bars :
    bar.set_hatch(patterns.pop(0))
# bars[0].set_hatch('/')
# bars[1].set_hatch('O')
# bars[2].set_hatch('*')
plt.figure(figsize=(6, 4))
plt.show()
```
